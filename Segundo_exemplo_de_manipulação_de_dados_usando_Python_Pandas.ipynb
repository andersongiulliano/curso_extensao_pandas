```python
import pandas as pd

man_df = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/man_veic.xlsx')
display(man_df.head())
```

```python
#Inserção simples de dado
man_df['Veículo consertado'] = True

display(man_df.head())
```

```python
# Criar uma coluna "Obs" que tenha nela escrito "MELHOR OFICINA" quando a coluna Oficina for igual a TDR CENTRO AUTOMOTIVO DE PECAS

import numpy as np

man_df['Obs'] = ["MELHOR OFICINA" if Oficina == 'TDR CENTRO AUTOMOTIVO DE PECAS' else "" for Oficina in man_df['Oficina']]

display(man_df.loc[man_df['Oficina'].isin(['TDR CENTRO AUTOMOTIVO DE PECAS','LIMPAUTO NATAL']), ['Oficina', 'Obs']])
```

```python
# Criar um dataframe finalizadas_df contendo 
# apenas linhas com situação FINALIZADA e exibir na tela
finalizadas_df = man_df.loc[man_df['Situacao'] == 'FINALIZADA']
display(finalizadas_df)
```

```python
# (por Leandro Rodrigues)
# como preencher uma coluna 'Status da NF' que verifica o seguinte:
# se o valor de venda for maior que 200 reais, ele fala que tá Caro..caso contrário, está barato

import numpy as np

finalizadas_df['Status da NF'] = np.where(finalizadas_df['ValorPecas'] > 200.00, 'Caro', 'Barato')
display(finalizadas_df[['Oficina', 'ValorPecas', 'Status da NF']])
```
