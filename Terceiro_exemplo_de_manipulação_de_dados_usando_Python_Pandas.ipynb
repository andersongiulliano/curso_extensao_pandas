```python
import pandas as pd

serv_df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/relatorio_serv.txt', sep=';')
display(serv_df)
```

```python
#Inserção simples de dado
serv_df['Data da NF'] = True
serv_df['ISS %'] = True
serv_df['VALOR ISS (R$)'] = True

display(serv_df.head())
```

```python
novo_serv_df = serv_df.drop(columns = ['Data Pgto. Órgão','Unidade','Serviço','Total Retido MDO','Desconto Residual','NF Peças','Modelo','Ano Veículo','Km','Cartão','Veículo Provisório','Data Pgto. Credenciado','Marca','Total Retido Peças','Carta de Correção','Empenho Peças','Empenho MDO','Processo','Unnamed: 35'])
display(novo_serv_df.head(2))



novo_serv_df.to_excel('relatorio_serviços.xlsx', sheet_name='COMPLETO', index='FALSE')
```

```python
# Convertendo strings em Float nas colunas para depois fazer a soma por subunidade
# Pode ser feito em uma linha:
# df['coluna'] = df['coluna'].str.replace(',','.').astype(float)

novo_serv_df['Valor Bruto'] = novo_serv_df['Valor Bruto'].str.replace(',','.')
novo_serv_df['Valor Considerado'] = novo_serv_df['Valor Considerado'].str.replace(',','.')
novo_serv_df['Valor Liquido'] = novo_serv_df['Valor Liquido'].str.replace(',','.')

novo_serv_df['Valor Bruto'] = novo_serv_df['Valor Bruto'].astype(float)
novo_serv_df['Valor Considerado'] = novo_serv_df['Valor Considerado'].astype(float)
novo_serv_df['Valor Liquido'] = novo_serv_df['Valor Liquido'].astype(float)





#novo_serv_df['Valor Bruto'] = novo_serv_df['Valor Bruto'].astype(float)
```

```python
# Soma dos gastos por subunidade dentro de um mês
soma_serv = novo_serv_df[['Subunidade','Valor Bruto','Valor Considerado','Valor Liquido']].groupby(by='Subunidade').sum().round(2)

# Acrescentar a linha total com a soma de cada coluna:
soma_serv.loc['TOTAL'] = soma_serv.sum() #(axis=0) == ()

# Se houvesse necessidade de somar as linhas: df1['Total'] = df1.sum(axis=1)

display(soma_serv)
```

```python
# Calcular a taxa de serviço em um nova coluna distribuindo proporcionalmente o valor por subunidade

taxa = float(input())

soma_serv['Taxa de Serviço'] = (soma_serv['Valor Considerado']/soma_serv['Valor Considerado'].iloc[-1]*taxa ).round(2)

display(soma_serv)
```
